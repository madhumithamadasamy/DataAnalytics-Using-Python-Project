```python
import pandas as pd
sales_data = {
    "Cus_nm": ["Ravi", "Anu", "Ravi", None, "Kumar"],
    "Sales": [500, "abc", 500, 700, None],
    "Cty": ["Chennai", "Mumbai", "Chennai", None, "Delhi"]
}
df = pd.DataFrame(sales_data)

print("Original Dataset:")
print(df)

print("\nChecking for Null Values:")
print(df.isnull())

clean_df=df.dropna()
print("\nCleaned Dataset:")
print(clean_df)

df.fillna(0,inplace=True)
print(df)

print("\n checking the duplicates:")
print(df.duplicated())

print("\n dropping the duplicates:")
print(df.drop_duplicates())

df.rename(columns={"Cus_nm":"Customer_Name","Cty":"City"},inplace=True)
print(df)
```
