# Data Filtering

```python
import pandas as pd
data={
    "Region":["South","North","East","West","South","North"],
    "Product":["Laptop","Mobile","Laptop","Tablet","Mobile","Tablet"],
    "Sales":[50000,30000,45000,25000,20000,35000]
}
df=pd.DataFrame(data)
print(df)

```

**filtering concept**

```python
south_sales = df[df["Region"] == "South"]
print(south_sales)
```

```python
high_sales=df[df["Sales"]>30000]
print(high_sales)
```

**Sorting concept**

```python
#Ascending order
sorted_df=df.sort_values("Sales")
print(sorted_df)
```

```python
#Descending order
sorted_df=df.sort_values("Sales",ascending=False)
print(sorted_df)
```

**Groupby concept** day-8

```python
region_sales=df.groupby("Region")["Sales"].sum()
print(region_sales)
```

```python
product_sales=df.groupby("Product")["Sales"].sum()
print(product_sales)
```

**Aggregation** day-8

```python
print(df["Sales"].sum())
```

```python
print(df["Sales"].mean())
```

```python
print(df["Sales"].max())
```

```python
print(df["Sales"].min())
```

```python
print(df["Sales"].count())
```

```python
result = df.groupby("Region")["Sales"].agg(
    ["sum", "mean", "max", "min", "count"]
)
print(result)
```
