```python
import numpy as np
products = ["Laptop","Mobile","Tablet","Headphone","Smartwatch"]

#Sales units
sales = np.array([120,250,80,150,100])

#Total sales
print("Total Sales:",np.sum(sales))

#Average sales
print("Average Sales:",np.mean(sales))

# Highest sales
print("Maximum Sales:",np.max(sales))

#Lowest sales
print("Minimum Sales:",np.min(sales))

#best-selling product
best_index = np.argmax(sales)
print("Best-selling product:",products[best_index])

#Product-wise sales
print("\n Product Sales Report")
for i in range(len(products)):
    print(products[i],":",sales[i])
```
