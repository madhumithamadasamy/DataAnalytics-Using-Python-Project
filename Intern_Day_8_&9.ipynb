# DATA VISUALIZATION (Matplotlib)

**Line chart**

```python
import matplotlib.pyplot as plt
months = ["Jan", "Feb", "Mar", "Apr","May"]
sales = [10000,12000,15000,13000,18000]
plt.plot(months,sales)
plt.title("Sales Report")
plt.xlabel("Months")
plt.ylabel("Sales Amount")
plt.show()
```

```python
#customaization
import matplotlib.pyplot as plt
months = ["Jan", "Feb", "Mar", "Apr","May"]
sales = [10000,12000,15000,13000,18000]
plt.plot(months,sales,color="red",marker="o",linestyle="--")
plt.title("Sales Report")
plt.xlabel("Months")
plt.ylabel("Sales Amount")
plt.show()
```

**Bar chart**

```python
import matplotlib.pyplot as plt
students = ["Alice", "Bob", "Charlie", "David"]
Marks = [85, 92, 78, 95]
plt.bar(students,Marks)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
#customization
import matplotlib.pyplot as plt
students = ["Alice", "Bob", "Charlie", "David"]
Marks = [85, 92, 78, 95]
plt.bar(students,Marks,color=["green","red","blue","purple"])
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

**Pie chart Day-9**

```python
import matplotlib.pyplot as plt
expenses = [5000,3000,2000]
categories = ["Food","Travel","Shopping"]
plt.pie(expenses,labels=categories)
plt.title("Expense Report")
plt.show()
```

```python
import matplotlib.pyplot as plt
expenses = [5000,3000,2000]
categories = ["Food","Travel","Shopping"]
plt.pie(expenses,labels=categories, autopct="%1.1f%%", colors=["red","purple","yellow"])
plt.title("Expense Report")
plt.show()
```

**Dashboard**-combine all charts

```python
import matplotlib.pyplot as plt
months = ["Jan","Feb","Mar","Apr","May","Jun"]
sales = [10000,12000,15000,13000,18000,20000]
plt.figure(figsize=(18,5))
#Line Chart
plt.subplot(1,3,1)
plt.plot(months,sales,color="blue",marker="o")
plt.title("Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales")

#Barchart
plt.subplot(1,3,2)
plt.bar(months,sales,color="orange")
plt.title("Sales comparison")

#Pie chart
plt.subplot(1,3,3)
expenses = [5000,3000,2000]
categories = ["Food","Travel","Shopping"]
plt.pie(expenses,labels=categories, autopct="%1.1f%%", colors=["pink","purple","blue"])
plt.title("Expense Report")
plt.tight_layout()
plt.show()
```

# ADVANCED VISUALIZATION





**HISTOGRAM**

```python
import matplotlib.pyplot as plt
marks = [45,50,55,60,65,70,72,75,80,85,90,95]
plt.hist(marks,bins=5)
plt.title("Student Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()
```

**SCATTER PLOTS**

```python
import matplotlib.pyplot as plt
hours = [1,2,3,4,5,6,7,8]
scores = [35,40,50,60,65,75,85,95]
plt.scatter(hours,scores)
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

**SUB PLOTS**

```python
import matplotlib.pyplot as plt
plt.subplot(1,2,1)
plt.plot([1,2,3,4],[10,20,30,40])
plt.title("Graph")

plt.subplot(1,2,2)
plt.plot([1,2,3],[30,20,10])
plt.title("Graph2")
plt.show()
```
